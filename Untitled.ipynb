```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
G = 9.8
```

```python
#Make a Function to define height
def height_calculator(v_0,t):
    return v_0*t-0.5*G*t**2
```

```python
v_0_list = np.linspace(1.0,100.0,10)
t = np.linspace(0,10)
for v_0 in v_0_list:
    h_func = np.vectorize(lambda t: height_calculator(v_0,t))
    h = h_func(t)
    plt.scatter(t,h)
plt.xlabel('time (s)')
plt.ylabel('height (m)')
plt.legend(v_0_list)
plt.ylim(bottom=0)
plt.show()
    
```
