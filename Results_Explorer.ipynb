```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from seaborn import scatterplot
```

```python
logs = []
logs.append(pd.read_csv('Resuts/SVC_S19_1.csv'))
logs.append(pd.read_csv('Resuts/SVC_S72_1.csv'))
logs.append(pd.read_csv('Resuts/SVC_S266_1.csv'))
(fig,axes)=plt.subplots(1,len(logs), figsize  = (5*len(logs),5))
best_rows=[]
for i,ax in enumerate(axes):
    data = logs[i]
    best_row = data[data['mean_test_score']==data['mean_test_score'].max()]
    best_rows.append(best_row)
    data['selection_factor'] = data['param_select_features__drt'] + data['param_select_features__npeaks']
    hue = data['param_select_features__group'].map(str)
    #hue = data['param_clf__C'].map(lambda i: 'C: '+str(i))
    sns.scatterplot(x=data['selection_factor'], y= data['mean_test_score'], hue = hue, ax=ax)
#plt.savefig('Resuts/LogReg__C.png', format='png')
plt.show()
```
